```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset['charges']
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
   independent,
    dependent,
    test_size=0.30,
    random_state=0
)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error

ada_pipeline = Pipeline([
    ('model', AdaBoostRegressor(
        base_estimator=DecisionTreeRegressor(max_depth=3),
        n_estimators=100,
        learning_rate=0.1,
        random_state=42
    ))
])

# Train
ada_pipeline.fit(X_train, Y_train)

# Predict
Y_pred_ada = ada_pipeline.predict(X_test)

# Evaluate
print("AdaBoost MSE:", mean_squared_error(Y_test, Y_pred_ada))


```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred_ada)
```

```python
r_score
```

```python
import pickle

filename = "finalized_model_AdaBoost.sav"

# Open file in write-binary mode and dump model
with open(filename, 'wb') as file:
    pickle.dump(ada_pipeline, file)
```

```python
loaded_model=pickle.load(open("finalized_model_AdaBoost.sav",'rb'))
```
